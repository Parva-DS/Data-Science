```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Drop Columns

df=df.drop(['Market Category','Popularity'],axis=1)
```

```python
# Handle missing values

df['Engine HP']=df['Engine HP'].fillna(df['Engine HP'].median())
```

```python
df['Engine Cylinders']=df['Engine Cylinders'].fillna(df['Engine Cylinders'].median())
```

```python
df['Number of Doors']=df['Number of Doors'].fillna(df['Number of Doors'].mean())
```

```python
df['Engine Fuel Type']=df['Engine Fuel Type'].fillna(df['Engine Fuel Type'].mode()[0])
```

```python
df.isnull().sum()
```

```python

# Columns where outliers should be DROPPED
drop_cols = ['Year', 'Engine HP', 'MSRP']

# Select numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Columns where outliers will be replaced with median
replace_cols = [c for c in num_cols if c not in drop_cols]

# Step 1: Calculate IQR for all numeric columns
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Create outlier mask
outlier_mask = ((df[num_cols] < (Q1 - 1.5 * IQR)) |
                (df[num_cols] > (Q3 + 1.5 * IQR)))

# Step 3: Replace outliers with median (except 3 columns)
for col in replace_cols:
    median_value = df[col].median()
    df.loc[outlier_mask[col], col] = median_value

# Step 4: Drop rows where Year, EngineHP, MSRP have outliers
df_clean = df[~outlier_mask[drop_cols].any(axis=1)].reset_index(drop=True)

df_clean
```

```python
# Number of unique brands and models

print(df_clean['Make'].nunique())
print(df_clean['Model'].nunique())
```

```python
# Top 10 car brands by model_count

top_makes=df_clean['Make'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_makes.index,y=top_makes.values)
plt.title("Top 10 Car Manufacturers by Model Count")
plt.xlabel("Car Brands")
plt.ylabel("Number of models")
plt.show()
```

```python
# Top 10 Manufacturers by average price

manufacturer=df_clean.groupby('Make')['MSRP'].mean().sort_values(ascending=False)
manufacturer.head(10)
```

```python
# Car brands based on Average price

luxury_threshold=50000
economy_threshold=30000

luxury_brands=manufacturer[(manufacturer>50000)]
economy_brands=manufacturer[(manufacturer<30000)]
mid_range_brands=manufacturer[(manufacturer<=50000) & (manufacturer>=30000)]

luxury_brand_count=len(manufacturer[manufacturer>luxury_threshold])
economy_brand_count=len(manufacturer[manufacturer<economy_threshold])
mid_range_brand_count=len(manufacturer[(manufacturer<=luxury_threshold) & (manufacturer>=economy_threshold)])

categories=['luxury_brands','economy_brands','mid_range_brands']
counts=[luxury_brand_count,economy_brand_count,mid_range_brand_count]
```

```python
plt.bar(categories,counts,color=['lightpink','lightgreen','lightblue'])
plt.title("Brand count by Categories")
plt.xlabel("Category")
plt.ylabel("Number of Brands")
```
