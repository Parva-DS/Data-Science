```python
class atm:
    def __init__(self):
        self.pin=''
        self.balance=0
        self.transaction_history=[]
        self.create_pin()
        

    def menu(self):
        user_input=input('''
                        Hello, Welcome to Apna Bank:
                        How may i help you?
                        1. Press 1 to deposit money
                        2. Press 2 to check balance
                        3. Press 3 to withdraw money
                        4. Press 4 to change pin
                        5. Press 5 to check the last few transaction
                        6. Press 6 to exit ''')
        
        if user_input=='1':
            self.deposit_money()
        elif user_input=='2':
            self.check_balance()
        elif user_input=='3':
            self.withdraw_money()
        elif user_input=='4':
            self.change_pin()
        elif user_input=='5':
            self.transaction_check()
        else:
            self.exit()

    def create_pin(self):
        pin_create=input("Enter 4 digit pin")
        self.pin=pin_create
        print("Your pin has been created")
        self.deposit_money()

    
    def deposit_money(self):
        user_pin=input("Enter ypur pin")
        if user_pin==self.pin:
            amount=int(input("Enter an amount you want to deposit"))
            self.balance+=amount
            self.transaction_history.extend([amount,"Deposited"])
            print(f"You deposited {amount} rupees in your bank account")
            self.menu()
        else:
            print("Invalid pin")
            self.menu()


    def withdraw_money(self):
        user_pin=input("Enter your pin")
        if user_pin==self.pin:
            amt=int(input("Enter an amount you want to withdraw from your account"))
            self.transaction_history.extend([amt,"Withdraw"])
            if amt>50000:
                print("You can withdraw money less than 50000")
            else:
                print(f"You withdraw {amt} rupees from your bank account")
                self.balance-=amt
            self.menu()
        else:
            print("Invalid pin")
            self.menu()


    def check_balance(self):
        user_pin=input("Enter your pin")
        if user_pin==self.pin:
            total_balance=self.balance
            print(f"Total Balance: {total_balance}")
            self.menu()
        else:
            print("Invalid pin")
            self.menu()


    def change_pin(self):
        user_pin=input("Enter your pin")
        if user_pin==self.pin:
            new_pin=input("Enter a new pin")
            self.pin=new_pin
            print("Your pin has been updated successfully")
            self.menu()
        else:
            print("Invalid pin")
            self.menu()


    def transaction_check(self):
        user_pin=input("Enter your pin")
        if user_pin==self.pin:
            print("*****LAST FEW TRANSACTION*****")
            if not self.transaction_history:
                print("No transaction")
            else:
                for i in self.transaction_history:
                    print("--",i)
            self.menu()
        else:
            print("Invalid pin")
            self.menu()


    def exit(self):
        print("You have exited your bank account")

a=atm()
```
